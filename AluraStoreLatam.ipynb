# **1. Análisis de facturación**


```python
import matplotlib.pyplot as plt

facturacion_total = [
    tienda['Precio'].sum(),
    tienda2['Precio'].sum(),
    tienda3['Precio'].sum(),
    tienda4['Precio'].sum()
]
nombres_tiendas = ['tienda', 'tienda 2', 'tienda 3', 'tienda 4']

plt.figure(figsize=(8,5))
plt.bar(nombres_tiendas, facturacion_total, color='skyblue')
plt.title('Facturación total por tienda')
plt.ylabel('Facturación ($)')
plt.xlabel('Tienda')
plt.show()

plt.figure(figsize=(6,6))
plt.pie(facturacion_total, labels=nombres_tiendas, autopct='%1.1f%%', startangle=90)
plt.title('Porcentaje de facturación por tienda')
plt.show()

```

Analisis facturacion

Para la facturacion, la tienda 1 representa mayor cantidad de ventas de productos, segun los graficos de barras y pastel, por lo que se reafirma que la tienda 1 es la que mas factura productos.

# 2. Ventas por categoría

```python
categorias = tienda['Categoría del Producto'].unique()
facturacion_categorias = {
    'Tienda 1': tienda.groupby('Categoría del Producto')['Precio'].sum(),
    'Tienda 2': tienda2.groupby('Categoría del Producto')['Precio'].sum(),
    'Tienda 3': tienda3.groupby('Categoría del Producto')['Precio'].sum(),
    'Tienda 4': tienda4.groupby('Categoría del Producto')['Precio'].sum(),
}
df_fact_cat = pd.DataFrame(facturacion_categorias).fillna(0)

df_fact_cat.T.plot(kind='bar', stacked=True, figsize=(10,6))
plt.title('Facturación por categoría y tienda')
plt.ylabel('Facturación ($)')
plt.xlabel('Tienda')
plt.legend(title='Categoría', bbox_to_anchor=(1.05, 1))
plt.tight_layout()
plt.show()
```

Analisis venta por categoria

segun el grafico de barras apiladas, podemos deducir que la categoria con mas ventas es la de Electronicos en las 4 tiendas, siendo la segunda favorita la categoria de Electrodomesticos.

## Informe

### Introducción

El propósito de este informe es identificar cuál de las cuatro tiendas actualmente operadas por el Sr. Juan (Tienda 1, Tienda 2, Tienda 3 y Tienda 4) debería vender, basándose en un análisis comparativo de desempeño comercial. Se evaluaron cinco criterios clave:

    -Facturación total

    -Ventas por categoría

    -Calificación promedio de clientes

    -Productos más y menos vendidos

    -Costo de envío promedio

Con esta información se busca tomar una decisión objetiva, enfocada en conservar las tiendas más rentables, eficientes y bien valoradas, descartando aquella con menor desempeño global.

### Facturación total
La Tienda 1 se posiciona como la más rentable en términos de facturación. Tanto los gráficos de barras como los de pastel indican que esta tienda tiene el mayor volumen de ventas, lo que la hace estratégica desde el punto de vista financiero. Su amplio movimiento de productos la convierte en un canal sólido de ingresos.

-Tienda 1: mantener.

### Ventas por categoría
La categoría de Electrónicos domina claramente en todas las tiendas, seguida por Electrodomésticos. No obstante, es relevante observar qué tan diversificadas están las ventas por tienda:

Si una tienda depende excesivamente de una sola categoría, podría tener riesgo operativo ante cambios de mercado.

Las tiendas que muestran una mayor variedad pueden ser más estables a largo plazo.

El análisis indica que todas siguen un patrón similar, por lo tanto este criterio no descarta claramente ninguna tienda, pero favorece a aquellas con ventas más diversificadas (como Tienda 3 y 4).

### Calificación promedio de clientes
Este indicador refleja la experiencia del usuario:

-Tienda 3 es la mejor valorada con 4.04 de calificación promedio.

-Le sigue Tienda 2 con 4.03.

Tienda 1 y Tienda 4 tienen puntuaciones menores.

Una diferencia marginal, pero suficiente para destacar que Tienda 3 tiene la mejor percepción del cliente, lo cual impacta la fidelización y repetición de compra.

-Tienda 3: mantener.

### Productos más y menos vendidos
Analizando los productos más vendidos:

Tienda 1: TV LED UHD 4K – 60 unidades

Tienda 2: Curso “Iniciando en programación” – 65 unidades

Tienda 3: Kit de bancas – 57 unidades

Tienda 4: Cama box – 62 unidades

El producto más vendido de todas las tiendas pertenece a la Tienda 2, lo cual es un punto fuerte.

Sin embargo, la falta de más productos populares en esta tienda podría significar una alta dependencia de un solo artículo, lo que limita su estabilidad.

### Costo de envío promedio
Tienda 4: el costo de envío más bajo, con $23,459.45

Tienda 1: el más alto, con $26,018.60

Esto influye directamente en la competitividad del precio final para los clientes. Tiendas con envío más barato tienen ventaja en plataformas donde el costo total define la conversión.

-Tienda 4: mantener.

### Conclusión: Tienda a vender – Tienda 2
Aunque la Tienda 2 cuenta con el producto más vendido, su desempeño global es menos robusto en comparación con las demás. Sus principales debilidades:

Calificación del cliente apenas ligeramente mejor que otras, pero no líder.

Alto grado de dependencia en un solo producto.

No destaca en facturación total.

No presenta ventaja logística (costo de envío no competitivo).

Sin diferenciación fuerte en categorías o diversidad de productos.

Por lo tanto, se recomienda al Sr. Juan vender la Tienda 2 y conservar las tiendas 1, 3 y 4, que presentan ventajas en ingresos, percepción del cliente y costos operativos.
